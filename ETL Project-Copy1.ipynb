```python
import pandas as pd
from sqlalchemy import create_engine
import json
import os
import pprint
from pandas.io.json import json_normalize
```

```python
#reading csv file and turning it into a dataframe
csv_file= "Restaurants_in_Wake_County.csv"
restaurant_df= pd.read_csv(csv_file)
restaurant_df.head()
```

```python
#cleaned by selecting columns needs
updated_restaurant_df= restaurant_df[['OBJECTID', 'NAME', 'ADDRESS1', 'CITY', 'STATE', 'POSTALCODE']]
updated_restaurant_df.head()
```

```python
updated_restaurant_df.rename(columns={'NAME':'RESTAURANTNAME',
                          'ADDRESS1':'ADDRESS'},
                 inplace=True)
updated_restaurant_df.set_index('OBJECTID')
```

```python
#read in json file 
filepath = os.path.join("Food_Inspection_filtered.json")
#json_file="Food_Inspections.json"
#inspection_df=pd.read_json(json_file)
#inspection_df.head()
with open (filepath, encoding="utf8") as jsonfile:
    inspection_json= json.load(jsonfile)

    inspection_json
```

```python
#turn into a dataframe using json_normalize
inspection_df=pd.DataFrame.from_dict(json_normalize(inspection_json), orient='columns')
inspection_df
#data needed in the first features column
```

```python
#unnest features column, turned into a dataframe
updatedinspection = json_normalize(inspection_json['features'])
updatedinspection
```

```python
#reduced columns needed
inspection_df= updatedinspection[['properties.OBJECTID', 'properties.SCORE', 'properties.TYPE', 'properties.DESCRIPTION']]
inspection_df.head()
```

```python
#changed column name
inspection_df.rename(columns={'properties.OBJECTID':'OBJECTID',
                          'properties.SCORE':'SCORE',
                          'properties.TYPE':'INSPECTIONTYPE',
                          'properties.DESCRIPTION':'DESCRIPTION'},
                 inplace=True)
inspection_df.set_index('OBJECTID')
inspection_df.head()
```
